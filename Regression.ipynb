## REMOVE YEAR

```python
data_clean = data_clean.drop(['Year'],axis = 1)
data_clean
```

## CREATE DUMMY VARIABLES

```python
data_clean = pd.get_dummies(data_clean,drop_first = True)
data_clean
```

## DATA SCALING

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
y = data_clean['Log_Price']
x = data_clean.drop(['Log_Price'],axis = 1)
scaler.fit(x)
inputs = scaler.transform(x)
```

## SPLIT TO TRAIN AND TEST

```python
from sklearn.model_selection import train_test_split as split
```

```python
x_train,x_test,y_train,y_test = split(inputs, y, train_size = 0.8, random_state = 365)
```

## DATA TRAINING

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

```python
yhat = reg.predict(x_train)
y = y_train
plt.scatter(yhat,y)
plt.xlabel('y_hat')
plt.ylabel('target')
plt.xlim(6,13)
plt.ylim(6,13)
```

## TESTING

```python
reg.fit(x_test,y_test)
```

```python
reg.score(x_test,y_test)
```

```python
y_hat_test = reg.predict(x_test)
y = y_test
plt.scatter(y_hat_test,y)
plt.xlabel('Prediction')
plt.ylabel('Target')
plt.xlim(6,13)
plt.ylim(6,13)
```

```python
y_hat_test = np.exp(y_hat_test)
```

```python
results_df = pd.DataFrame(columns = ['Prediction'], data = y_hat_test)
y = np.exp(y).reset_index(drop = True)
results_df['Target'] = y 
results_df.reset_index(drop = True)
results_df
```

```python
results_df['Residuals'] = (results_df['Target']-results_df['Prediction'])
results_df
```

```python
results_df['Difference in %'] = np.absolute(results_df['Residuals'])/results_df['Target']
results_df
```

```python
results_df.describe()
```

```python
results_df = results_df.sort_values(by = ['Difference in %'])
results_df
```

```python
results_df.head()
```

```python
results_df.tail()
```

## GET ACCURACY OF PREDICTION

```python
results_df["Accuracy%"] = 100 - results_df['Difference in %']
results_df
```

```python
Accuracy = results_df['Accuracy%'].mean()
print('Prediction Accuracy is', Accuracy,'%')
```
